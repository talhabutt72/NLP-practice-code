```python
sentence = """
Natural language processing (NLP) is the processing of natural language information by a computer. NLP is a subfield of computer science and is closely associated! with artificial intelligence. NLP is also related to information retrieval, knowledge representation, computational linguistics, and linguistics more broadly.
"""
```

```python
from nltk.tokenize import sent_tokenize, word_tokenize, WhitespaceTokenizer

tokenize = sent_tokenize(sentence)
tokenize
```

```python
word_toknizer = word_tokenize(sentence)
word_toknizer
```

```python
tokenizer = WhitespaceTokenizer()
space_tokenizer = tokenizer.tokenize(sentence)
space_tokenizer
```

```python
words = ["talks", "studying", "studies", "happiness", "organization", "relational", "universities"]
```

```python
# Stemming

# PorterStemming

from nltk.stem import PorterStemmer, LancasterStemmer, SnowballStemmer

stem_porter = PorterStemmer()

for word in words:
  print(word + "-----> " + stem_porter.stem(word))
```

```python
lan_stem = LancasterStemmer()

for word in words:
  print(word + "-----> " + lan_stem.stem(word))
```

```python
ball_stem = SnowballStemmer(language = "english")

for word in words:
  print(word + "-----> " + ball_stem.stem(word))
```

```python
from nltk.stem import WordNetLemmatizer

lemm = WordNetLemmatizer()

for word in words:
  print(word + "-----> " + lemm.lemmatize(word,'v'))
```
